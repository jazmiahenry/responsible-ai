```python
from transformers import AutoTokenizer, BloomModel
import torch

tokenizer = AutoTokenizer.from_pretrained("bigscience/bloom-560m")
model = BloomModel.from_pretrained("bigscience/bloom-560m")

inputs = tokenizer("Hello, my dog is cute", return_tensors="pt")
outputs = model(**inputs)

last_hidden_states = outputs.last_hidden_state
```

```python
from torchviz import make_dot
make_dot(last_hidden_states.mean(), params=dict(model.named_parameters()))
```

```python
make_dot(last_hidden_states.mean(), params=dict(model.named_parameters())).render("attached", format="png")
```
